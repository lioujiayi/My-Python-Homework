```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 畫一張圖

```python
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
```

# 接力畫一張完整的圖

```python
plt.figure(figsize=(6,6))

plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.legend('lower left')#將圖標放在左下角
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
#設定x軸名字
plt.xlim(0,10)
plt.ylim(-1,1)
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(-1,1,0.5))
plt.title('Sin and Cos')
```

# Variable Length Function Arguments

```python
def f(x):
    print(x)
```

```python
f('Hello')
```

```python
F('Hello','world')
```

```python
def ARGV(*argv):
    for i in argv:
        print(i)
```

```python
ARGV('Hello','world')
```

```python
ARGV('Hello','world','I','am','hungry')
```

```python
def ARGV1(arg1,*argv):
    print(arg1)
    print('')
    for i in argv:
        print(i)
```

```python
ARGV1('Hello','world','I','am','hungry')
```

```python
def ARGV2(*argv,arg1):
    print(arg1)
    print('')
    for i in argv:
        print(i)
```

```python
ARGV2('Hello','world','I','am','hungry',arg1='YA')
```

# Variable Length Keyword Argument

```python
def KWANG(**kwargs):
    for key,value in kwargs.items():
        print(f'{key}=={values}')
```

```python
KWANG(first="N",second="C",third="T",fourth="S")
```

```python
def Four(arg1,arg2,arg3,arg4):
    print('arg1 is ',arg1)
    print('arg2 is ',arg2)
    print('arg3 is ',arg3)
    print('arg4 is ',arg4)
```

```python
args=('N','C','T','S')
```

```python
Four(*args)
```

```python
kwargs={'arg2':"T",'arg1':"N",'arg4':"S",'arg3':"C"}
```

```python
Four(**kwargs)
```
