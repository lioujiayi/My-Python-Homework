# 各年度國產及進口料理米酒數量回歸分析

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('https://www.nta.gov.tw/opendata/D6-11_%E5%9C%8B%E7%94%A2%E5%8F%8A%E9%80%B2%E5%8F%A3%E6%96%99%E7%90%86%E7%B1%B3%E9%85%92%E6%95%B8%E9%87%8F%E8%A1%A8.csv')
df.head()
```

```python
df['年月']=df.年度*100+df.月份*100/12
```

```python
df.head()
```

```python
x=np.array(df.年月)
X=x.reshape(len(x),1)
```

```python
myxticks=[]
for i in range(len(df)):
    myxticks.append(str(df.年度[i])+'/'+str(df.月份[i]))
```
